```python
avSafety['incidents_85_99'] = avSafety['incidents_85_99']/avSafety['incidents_85_99'].max()
avSafety['fatal_accidents_85_99'] = avSafety['fatal_accidents_85_99']/avSafety['fatal_accidents_85_99'].max()
avSafety['fatalities_85_99'] = avSafety['fatalities_85_99']/avSafety['fatalities_85_99'].max()
temp= avSafety.iloc[:,2:5]
col_list= ['incidents_85_99','fatal_accidents_85_99','fatalities_85_99']

temp['total']=temp[col_list].sum(axis=1)
temp.head(5)
```

```python
avSafety.dropna(subset = ["incidents_85_99", "fatal_accidents_85_99", "fatalities_85_99"],axis = 0, inplace = True)
```

```python
avSafety[["incidents_85_99", "fatal_accidents_85_99", "fatalities_85_99"]] = avSafety[["incidents_85_99", "fatal_accidents_85_99", "fatalities_85_99"]].astype("int", copy = True)
```

```python
categoria1 = np.linspace(min(avSafety['fatal_accidents_85_99']), max(avSafety['fatal_accidents_85_99']),4)
categoria1
```

```python
severidad=["Leve", "Medio", "Moderado"]
avSafety["Severidad"] = pd.cut(avSafety["fatal_accidents_85_99"], categoria1, labels = severidad, include_lowest = True)
avSafety[["fatal_accidents_85_99", "Severidad"]].head(15)
```

```python
avSafety["Severidad"].value_counts()
```
